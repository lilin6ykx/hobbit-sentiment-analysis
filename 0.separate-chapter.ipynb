```python
import re
import os

from google.colab import drive
try:
    drive.flush_and_unmount()
except Exception:
    pass

drive.mount('/content/drive', force_remount=True)

def split_chapters_from_file(input_file="/content/drive/MyDrive/digphil/the-hobbit.txt", output_dir="chapters"):
    # Read the whole book into text
    with open(input_file, "r", encoding="utf-8") as file:
        book_text = file.read()
    # Create output directory if not exists
    os.makedirs(output_dir, exist_ok=True)
    # Compile a regex pattern for chapter starts (case-insensitive)
    chapter_pattern = re.compile(r'(?i)(chapter\s+\d+)', re.MULTILINE)
    starts = [m.start() for m in chapter_pattern.finditer(book_text)]
    chapters = []
    for i in range(len(starts)):
        start = starts[i]
        end = starts[i+1] if i+1 < len(starts) else len(book_text)
        chapters.append(book_text[start:end].strip())
    # Write each chapter to a file
    for idx, content in enumerate(chapters, 1):
        filename = os.path.join(output_dir, f"chapter_{idx}.txt")
        with open(filename, "w", encoding="utf-8") as f:
            f.write(content)
    print(f"{len(chapters)} chapters saved to '{output_dir}'.")

# Run this function to process "the-hobbit.txt"
# split_chapters_from_file()
split_chapters_from_file(input_file="/content/drive/MyDrive/digphil/the-hobbit.txt", output_dir="/content/drive/MyDrive/digphil/chapters")
```
